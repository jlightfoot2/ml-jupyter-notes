```python
# http://gluon.mxnet.io/chapter01_crashcourse/ndarray.html
# https://mxnet.apache.org/tutorials/index.html
import mxnet as mx
from mxnet import nd
mx.random.seed(1)
# just looking at shapes


# empty array no init
x = nd.empty((3, 4))
print(x)

# empty but with with init

x = nd.zeros((3, 4), mx.cpu(0))
print(x)
x = nd.ones((3, 4), mx.cpu(0))
print(x)


# random with normal distribution and zero mean
print("\nshape = (3,4)");
y = nd.random_normal(0, 1, shape=(3, 4))
print(y);

print("\nshape = 3");
# random with normal distribution and zero mean - different shape
y2 = nd.random_normal(0, 1, shape=3)
print(y2);

print("\nshape = (3,1)");
# random with normal distribution and zero mean - different shape #2
y3 = nd.random_normal(0, 1, shape=(3,1))
print(y3);

print("\nshape = (1,3)");
# random with normal distribution and zero mean - different shape #3
y4 = nd.random_normal(0, 1, shape=(1,3))
print(y4);
print();

# convert ndarray to numpy

yn = y.asnumpy()
print(yn)
print(type(yn))

```

```python
# get array metadata/info

#shape and size
print(y.shape)
print(y.size)
```

```python
# ndarray Operations
print(x)
print(y)
z = nd.ones(shape=(4,), ctx=mx.cpu(0))+4 
print(z)
#mult and addition
print(x + y) 
print(x*y) # element-wise multiplication
# addition with different shapes
print(y + z) 

# max
print("\n y max axis 1")
print(nd.max(y,axis=1))
print("\n y max axis 0")
print(nd.max(y,axis=0))

# exponentiation?
print("\n\n y exponentiation \n")
print(nd.exp(y))

# dot product
print("\ny orignal\n")
print(y)
print("\ny is transposed\n")
print(y.T)

print("\nMultiplication of\n")
print(x,y.T)

print("\nDot Product of x and y.T == \n")
print(nd.dot(x, y.T)) #y is transposed, See https://en.wikipedia.org/wiki/Transpose
```

```python
# data sum
data = nd.array([[[1,2],[2,3],[1,3]],
        [[1,4],[4,3],[5,2]],
        [[7,1],[7,2],[7,3]]])
print(data.shape)
print(data)
print("\naxis 2")
print(nd.sum(data, axis=2))
print("\naxis 1")
print(nd.sum(data, axis=1))
print("\naxis 0")
print(nd.sum(data, axis=0))
print("\naxis NULL")
print(nd.sum(data))
```

```python
# in-place operations


#show how memory address changes

print('id(y):', id(y))
y = y + x
print('id(y):', id(y))

# use [:] syntax to prevent new address

print('id(y):', id(y))
y[:] = x + y
print('id(y):', id(y))

# event better with memory

nd.elemwise_add(x, y, out=y)
```

```python
# Slicing

print(x)
print(x[1:3])
print(x)
x[1,2] = 9.0
print(x)
```

```python
# Broadcasting
# see http://gluon.mxnet.io/chapter01_crashcourse/ndarray.html#Broadcasting

x = nd.ones(shape=(3,3))
print('x = ', x)
y = nd.arange(3)
print('y = ', y)
print('x + y = ', x + y)

# Broadcast prefers to duplicate along the left most axis
# axis definition https://docs.scipy.org/doc/numpy-1.13.0/glossary.html
# http://gluon.mxnet.io/chapter01_crashcourse/linear-algebra.html#Length,-dimensionality,-and,-shape

y = y.reshape((3,1))
print('y = ', y)
print('x + y = ', x+y)
```

```python
# context
#http://gluon.mxnet.io/chapter01_crashcourse/ndarray.html#Managing-context

# run on first gpu
# z = nd.ones(shape=(3,3), ctx=mx.gpu(0)) #doesn't work on osx without cuda and built from source
# run on first cpu
z = nd.ones(shape=(3,3), ctx=mx.cpu(0)) 
print(z)

# copy var and its context

x_cpu = x.copyto(mx.cpu(0))
print(x_cpu)

# show context

print(x_cpu.context)
print(z.context)

#!!! In order to perform an operation on two ndarrays x1 and x2, we need them both to live on the same context
```

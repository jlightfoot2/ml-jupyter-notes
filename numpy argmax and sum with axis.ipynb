```python
import numpy as np
```

```python
a = np.arange(6).reshape(2,3)

b = np.arange(5,-1,-1).reshape(2,3)
f = np.array([[0,5,1],[4,2,3]]).reshape(2,3)
print("a val")
print(a)
print('no axis')
print(np.argmax(a))
print('axis 0')
print(np.argmax(a, axis=0))
print('axis 1')
print(np.argmax(a, axis=1))
print()
print("b val")
print(b)
print('no axis')
print(np.argmax(b))
print('axis 0')
print(np.argmax(b, axis=0))
print('axis 1')
print(np.argmax(b, axis=1))
print("f val")
print(f)
print('no axis')
print(np.argmax(f))
print('axis 0')
print(np.argmax(f, axis=0))
print('axis 1')
print(np.argmax(f, axis=1))
print()
# 3 dimensions

c = np.arange(12).reshape(2,3,2)

d = np.arange(12,0,-1).reshape(2,3,2)
print()
print("c val")
print(c)
print('no axis')
print(np.argmax(c))
print('axis 0')
print(np.argmax(c, axis=0))
print('axis 1')
print(np.argmax(c, axis=1))
print()
print("d val")
print(d)
print('no axis')
print(np.argmax(d))
print('axis 0')
print(np.argmax(d, axis=0))
print('axis 1')
print(np.argmax(d, axis=1))

# summing along axis
print("")
print("Summing")
print("a val")
print(a)
print('no axis')
print(np.sum(a))
print('axis 0')
print(np.sum(a, axis=0))
print('axis 1')
print(np.sum(a, axis=1))
print()
print("b val")
print(b)
print('no axis')
print(np.sum(b))
print('axis 0')
print(np.sum(b, axis=0))
print('axis 1')
print(np.sum(b, axis=1))
print("f val")
print(f)
print('no axis')
print(np.sum(f))
print('axis 0')
print(np.sum(f, axis=0))
print('axis 1')
print(np.sum(f, axis=1))
print()
```
